**MNIST**

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import scienceplots

mpl.rcParams.update({
    'font.size': 18,
    'axes.titlesize': 18,
    'axes.labelsize': 18,
    'xtick.labelsize': 18,
    'ytick.labelsize': 18,
    'legend.fontsize': 18,
})

df1 = pd.read_csv('metrics_DP_LP_mnist.csv')
df2 = pd.read_csv('metrics_DP_Static_mnist.csv')

rondas = df1['Round']
accuracies_lp = df1['Accuracy (%)']
epsilon_cummulative_lp= df1['Epsilon Acumulado']

accuracies_static = df2['Accuracy (%)']
epsilon_cummulative_static = df2['Epsilon Acumulado']

plt.style.use(['science', 'no-latex'])
fig, ax1 = plt.subplots(figsize=(8, 5))

color1 = 'C1'
color2 = 'C0'
ax1.set_xlabel('Rondas')
ax1.set_ylabel('Accuracy (%) / Epsilon', color='black')
ln1 = ax1.plot(rondas, accuracies_lp, color=color1, label='Accuracy (%)', linewidth=2)
ln2 = ax1.plot(rondas, epsilon_cummulative_lp, color=color1, label='Epsilon acumulado', linestyle='--', linewidth=2)
ln3 = ax1.plot(rondas, accuracies_static, color=color2, label='Accuracy (%)', linewidth=2)
ln4 = ax1.plot(rondas, epsilon_cummulative_static, color=color2, label='Epsilon acumulado', linestyle='--', linewidth=2)
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_ylim(0, 100)

lines = ln1 + ln3
labels = ["PD Optimizada", "PD Estatica"]
ax1.legend(lines, labels, loc='best')

fig.tight_layout()
plt.grid(True)
plt.savefig('LabelFlipping_MNIST.png', dpi=300)
```
